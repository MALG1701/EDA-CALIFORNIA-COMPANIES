```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#exportamos la data de una serie de empresas de USA

companies = pd.read_excel(r'C:\Users\mlopez\Downloads\1 - US COMPANY DATABASE SAMPLE.xlsx')
companies.head()
```

```python
#verificamos la info de la data

companies.info()
```

```python
#revisamos las metricas mas importantes de la data como cantidad, promedio, desviacion estandar y mas

companies.describe()
```

```python
#vemos el detalle de la data por columnas

companies.info
```

```python
#organizamos la data por la cantidad de ventas que han tenido para encontrar a la empresa con mas ventas y la cantidad

companies.sort_values(by=['company_exact_sales'], inplace=True, ascending=False)
max_sales=companies['company_exact_sales'].max()
max_sales
```

```python
#identificamos cual es la empresa con mas ingresos

companies_name=companies['company_name'].value_counts()
comp_more_sales=companies_name.index[0]

print('La empresa con mas ventas es:',comp_more_sales,'la cual ha vendido la cantidad de:',max_sales,'USD')
```

```python
#Organizamos las empresas segun sus ventas

companies_bysales=companies['company_sales'].value_counts()
companies_bysales
```

```python
#Graficamos las empresas y su rango de ingresos

x=companies_bysales.index
y=companies_bysales.values
plt.xticks(rotation=90)
plt.title('Ingresos de US Companias')
plt.ylabel('Cantidad de Companias')
plt.bar(x,y)
```

```python
#buscamos la cantidad de empleados que hay por empresas

companies_byemployees=companies['employee_range'].value_counts()
companies_byemployees
```

```python
#Graficamos los rangos de cantidad de empleados por empresa

a=companies_byemployees.index
b=companies_byemployees.values
plt.xticks(rotation=90)
plt.title('Cantidad de empleados US Companies')
plt.ylabel('Cantidad de Empleados')
plt.bar(a,b)
```

```python
#Ahora buscamos los ingresos por ciudad, creamos un nuevo dataframe

companies_city=companies[['city','company_exact_sales']].copy()
companies_city['city'] = companies_city['city'].fillna('')
companies_city['company_exact_sales'] = companies_city['company_exact_sales'].fillna(0)
companies_city
```

```python
#verificamos cuanta data tenemos por ciudad

companies_city['city'].value_counts()
```

```python
#Calculamos el ingreso promedio por empresas segun cada ciudad

avg_city_sales=companies_city.groupby('city')['company_exact_sales'].mean()
avg_city_sales
```

```python
#Graficamos las 10 primeras ciudades y su promedio de ingresos

c=avg_city_sales.index[1:10]
d=avg_city_sales.values[1:10]
plt.xticks(rotation=90)
plt.title('Promedio Ingresos Companias California')
plt.ylabel('USD')
plt.bar(c,d)
```
